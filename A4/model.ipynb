```python
import pandas as pd
import numpy as np
from sklearn import tree
df1 = pd.read_csv('Dataset/gender_submission.csv')
df2 = pd.read_csv('Dataset/test.csv')
df3 = pd.read_csv('Dataset/train.csv')
```

```python
survived = pd.DataFrame(df1)
train = pd.DataFrame(df2)
test = pd.DataFrame(df3)
train = train.drop(columns = ["Name"])
train = train.drop(columns = ["Ticket"])
for i in train["Sex"]:
    if i == "male":
        train = train.replace("male", 1)
    else:
        train = train.replace("female", 0)
        
train["Age"] = train["Age"].fillna(0)
train["Embarked"] = train["Embarked"].replace("Q", 1)
train["Embarked"] = train["Embarked"].replace("S", 2)
train["Embarked"] = train["Embarked"].replace("C", 3)
train = train.drop(columns = "Cabin")
train.loc[152, "Fare"] = 0
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train, survived)
```

```python
for i in test["Sex"]:
    if i == "male":
        test = test.replace("male", 1)
    else:
        test = test.replace("female", 0)
        
test["Age"] = test["Age"].fillna(0)
test["Embarked"] = test["Embarked"].replace("Q", 1)
test["Embarked"] = test["Embarked"].replace("S", 2)
test["Embarked"] = test["Embarked"].replace("C", 3)

test.fillna(0)
```

```python
import pickle
saved_model = pickle.dumps(clf)
pickle.dump(clf, open('model.pkl', 'wb'))
```
